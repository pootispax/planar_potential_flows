```python
%matplotlib inline
import numpy as np
import matplotlib.pylab as plt
import matplotlib.ticker as ticker
```

## Visualization of the elements of a matrix and orientation of the axes

Let's define a matrix `M`

```python
M = np.array([[1, 1.5, 2],
              [2, 2.5, 3]])
```

The `plt.imshow()` command is used to represent the components `M[i,j]` of the matrix in the form of a color map. 

```python
plt.imshow(M,cmap='coolwarm')
plt.colorbar();
```

The x and y axes can be modified to show integer indices

```python
plt.imshow(M,cmap='coolwarm')
tick_spacing = 1
ax = plt.gca()              # gca = get current axis, on peut utiliser alternativement: fig, ax = plt.subplots()
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
```

The option `origin='lower'` moves the origin of the `i` axis (line number) to the bottom. This axis then points upwards.

```python
ax = plt.gca()
plt.imshow(M,cmap='coolwarm',origin='lower')
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
```

If the matrix represents data from a grid covering a certain physical domain of dimension $L_x \times L_y$, the axes can be modified to correspond to this domain.

```python
x_max = 30.0
y_max = 20.0
bornes = [0, x_max, y_max, 0]   #[left, right, bottom, top]

plt.imshow(M,cmap='coolwarm',extent=bornes);
```

Let's improve this representation by moving the x-axis to the top of the figure.

```python
plt.imshow(M,cmap='coolwarm',extent=bornes)
ax = plt.gca()
ax.xaxis.tick_top()

#Ajout de flèches pour bien montrer les orientations des axes
ax.annotate('x', xy=(0, 0), xycoords='data', xytext=(35, 0.3), 
            arrowprops=dict(arrowstyle="<-", color='b',lw=3),size=15)
ax.annotate('y', xy=(0, 0), xycoords='data', xytext=(-0.1, 25), 
            arrowprops=dict(arrowstyle="<-", color='b',lw=3),size=15)

plt.arrow( 2.5, 2.5, 5, 0, length_includes_head = True, head_width = 1.0, lw=2, color='orange' )
ax.annotate('$j$', xy=(8, 2.9),size=20, color='orange')
plt.arrow( 2.5, 2.5, 0, 5, length_includes_head = True, head_width = 1., lw=2, color='orange' );
ax.annotate('$i$', xy=(2.8, 9),size=20, color='orange');
```

In this representation: <br>
x-axis <=> matrix axis <span style="font-family: monospace; ">j</span><br>
y-axis <=> matrix axis <span style="font-family: monospace; ">i</span>  (oriented downwards)
<div>

If we want the y-axis to point upwards, we can use

```python
bornes2 = [0, x_max, 0, y_max]   #[left, right, bottom, top]
plt.imshow(M,cmap='coolwarm',extent=bornes2,origin='lower');

#Ajout de flèches pour bien montrer les axes
ax = plt.gca()
ax.annotate('x', xy=(0, 0), xycoords='data', xytext=(35, -0.3), 
            arrowprops=dict(arrowstyle="<-", color='b',lw=3),size=15)
ax.annotate('y', xy=(0, 0), xycoords='data', xytext=(-0.1, 25), 
            arrowprops=dict(arrowstyle="<-", color='b',lw=3),size=15)

plt.arrow( 2.5, 2.5, 5, 0, length_includes_head = True, head_width = 1.0, lw=2, color='orange' )
ax.annotate('$j$', xy=(8, 2.9),size=20, color='orange')
plt.arrow( 2.5, 2.5, 0, 5, length_includes_head = True, head_width = 1., lw=2, color='orange' );
ax.annotate('$i$', xy=(3.5, 7.5),size=20, color='orange');
```

# Function np.meshgrid()

The `meshgrid()` function allows to create a grid of points.

```python
X = np.linspace(-2.0, 2.0, 5)
Y = np.linspace(-2.0, 2.0, 5)
XX, YY = np.meshgrid(X, Y)
plt.plot(XX,YY,ls='',marker='x',c='r',mew=2);
```

`XX` is a matrix containing the x-coordinates of all points on the grid. <br>
`YY` is a matrix containing the y-coordinates of all points on the grid.

```python
print('X =', X)
print('XX=\n', XX)
```

```python
print('Y =', Y)
print('YY=\n', YY)
```

Note: When `plt.plot()` receives matrices as arguments (as above), each column of the matrices is considered as a data set (a curve).

```python
#Let us show the various curves seen by the function `plot()`
plt.plot(XX,YY,linestyle='--',marker='x',c='r',mew=2);
```

<div style="margin-top: 0px; margin-left:0px; margin-right:auto; background-color:#C1FFC1;font-size: 120%;">Coordinate arrays created by the command <span style="font-family: monospace; ">meshgrid()</span> are particularly useful for evaluating a function on a grid.</div>

```python
ZZ = XX**2 + YY    #<-- Calculation of function f(x,y)=x**2 + y at each point of a grid
print(ZZ)
```

Plot of the ZZ array and of the corresponding grid

```python
plt.imshow(ZZ,cmap='coolwarm',extent=[-2.5, 2.5, -2.5, 2.5],origin='lower');
plt.plot(XX,YY,ls='',marker='x',c='r',mew=2);
```

Representation of the same function, but with a finer grid and adding some contours.

```python
X = np.linspace(-2.0, 2.0, 50)
Y = np.linspace(-2.0, 2.0, 50)
XX, YY = np.meshgrid(X, Y)
ZZ = XX**2 + YY
plt.imshow(ZZ,cmap='coolwarm',extent=[-2., 2., -2., 2.],origin='lower');
plt.contour(X,Y,ZZ);
```

```python
x_max = 30.0
y_max = 20.0
bornes2 = [0, x_max, 0, y_max]   #[left, right, bottom, top]

h = 10
X = np.linspace(0, x_max - h, 3)    #X = [0, 10 et 20]
Y = np.linspace(0, y_max - h, 2)    #Y = [0, 10]
XX, YY = np.meshgrid(X, Y)

#Décalons la grille (pour que les points soient au centre des cases)
XX = XX + h/2
YY = YY + h/2

plt.imshow(M,cmap='coolwarm',extent=bornes2,origin='lower');
plt.plot(XX,YY,linestyle='',marker='x',c='r',markersize=15,mew=2);
```

### Représentation d'un champ vectoriel

```python
vx = [[1, 1, 1],
     [2, 2, 2]]
vy = [[0, 0, 0],
     [2, 2, 2]]
plt.imshow(M,cmap='coolwarm',extent=bornes2,origin='lower');
plt.quiver(XX, YY, vx, vy, pivot="middle");
```
